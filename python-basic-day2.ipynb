```python
import random


stay = 0
switch = 0

doors = [1,0,0]
random.shuffle(doors)

for i in range(1000):
    choice  = random.randrange(3)
    if choice == 0:
        stay += 1
    else:
        switch += 1
        
print("stay: " + str(stay/1000))
print("switch:" + str(switch/1000))
```
